```python
 #Guess the number game
    
import random
answer = random.randint(10,20)
print("I am thinking of a number between 10 and 20")
i = 1
while True:
    enteredValue = int(input("Take a guess.\n "))
    if enteredValue == answer:
        print("Good job")
        break
    elif enteredValue >answer:
        print("Your guess is too high")
    elif enteredValue < answer:
        print("Your guess is too low")
    else:
        print("Please check the input")
    i+=1

print("You guessed the right number in " + str(i)+ " guesses")
```
